## Create unified Dataset

```python
dfs = {'abstract': abstract,
       'article': article,
       'blog': blog,
       'movie': movie,
       'reddit': reddit,
       'song': song,
       'twitter': twitter}

columns_to_keep = {'abstract': 'abstracts',
                   'article': 'content',
                   'blog': 'text',
                   'movie': 'Scripts',
                   'reddit': 'clean_comment',
                   'song': 'lyrics',
                   'twitter': 'clean_text'}  # Specify the column names to keep for each DataFrame

combined_data = pd.DataFrame(columns=['text', 'Source'])  # Initialize the combined DataFrame

for source, df in dfs.items():
    selected_column = df[columns_to_keep[source]]  # Select the specific column from the DataFrame
    combined_data = pd.concat([combined_data, pd.DataFrame({'text': selected_column, 'Source': source})], ignore_index=True)
```

```python
combined_data.to_csv('data.csv', index=False)
```

## Balancing the dataset

Targat, find the smallest group of sources and bring the others down to the same amount. 
We do not want the model to tend towards a category just because in the training phase more data from a specific category is being feed to the model


```python
# loading the complex data generated above
data_origin = pd.read_csv('Data\data.csv')
```

```python
# Calculate the minimum count among the sources
min_count = data_origin.groupby('Source').size().min()
min_count
```

```python
grouped_data = data_origin.groupby('Source')
grouped_data
```

```python
# Initialize an empty list to store the balanced data
balanced_data = []
grouped_data = data_origin.groupby('Source')

for source, group in grouped_data:
    sampled_subset = group.sample(n=min_count, random_state=42)
    balanced_data.append(sampled_subset)

# Concatenate the balanced data from all sources
balanced_data = pd.concat(balanced_data)
# Reset the index of the balanced data
balanced_data = balanced_data.reset_index(drop=True)

# The 'balanced_data' DataFrame now contains exactly 3923 observations for each source variation
# with the 'text' and 'Source' features preserved
```
